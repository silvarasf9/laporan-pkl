# Pendidikan

## Tahun 2023 Semester 1

```python
pendidikan_s1_23 = '/content/drive/My Drive/PKL/AGREGAT_202301/AGR_PENDIDIKAN_UMUR_TUNGGAL_3318_KAB. PATI.xlsx'
df_pendidikan_s1_23 = pd.read_excel(pendidikan_s1_23)
df_pendidikan_s1_23.columns
```

```python
pddk_s1_23 = df_pendidikan_s1_23.copy()

# Mengelompokkan kolom berdasarkan L (Laki-laki) dan P (Perempuan)
# Mengambil nama kategori tanpa akhiran L dan P
categories = [col[:-2] for col in pddk_s1_23.columns if col.endswith(' L') or col.endswith(' P')]
data = []

# Mengumpulkan data ke dalam format yang lebih terstruktur
for category in categories:
    # Memeriksa apakah kolom ada sebelum mengaksesnya
    laki_laki_col = category + ' L'
    perempuan_col = category + ' P'

    if laki_laki_col in pddk_s1_23.columns and perempuan_col in pddk_s1_23.columns:
        # Mengelompokkan berdasarkan umur
        for age in pddk_s1_23['UMUR'].unique():
            # Filter DataFrame berdasarkan umur
            filtered_data = pddk_s1_23[pddk_s1_23['UMUR'] == age]
            laki_laki_value = filtered_data[laki_laki_col].sum()  # Jumlah untuk Laki-laki
            perempuan_value = filtered_data[perempuan_col].sum()  # Jumlah untuk Perempuan

            data.append({'Pendidikan': category, 'UMUR': age, 'Gender': 'Laki-laki', 'Jumlah': laki_laki_value})
            data.append({'Pendidikan': category, 'UMUR': age, 'Gender': 'Perempuan', 'Jumlah': perempuan_value})
    else:
        print(f"Kolom tidak ditemukan: {laki_laki_col} atau {perempuan_col}")


pend_s1_23 = pd.DataFrame(data)
pend_s1_23
```

```python
pend_s1_23.describe(include='all')
```

```python
pend_s1_23.describe(include='object')
```

### plot 1

```python
# Membuat scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(pend_s1_23, x='UMUR', y='Jumlah', hue='Pendidikan', style='Gender', s=100, palette='deep')

plt.title('Keterkaitan antara Umur, Jumlah, dan Pendidikan Semester Pertama Tahun 2023')
plt.xlabel('Umur')
plt.ylabel('Jumlah')
plt.legend(title='Pendidikan dan Gender')
plt.grid(True)
plt.tight_layout()
plt.show()
```
